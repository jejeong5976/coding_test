## 골드바흐의 추측

```python
# 함수를 통해 매번 prime을 확인하는 것은 불필요한 중복 - 시간 초과
import math
import sys

MAX = 1000000
prime = [True] * (MAX + 1)
prime[1]=False
for i in range(2,int(math.sqrt(MAX))+1):
    if prime[i]:
        for j in range(i*i,MAX+1,i):
            prime[j]=False

while True:
    n=int(sys.stdin.readline())
    if n==0:
        break
    
    flag = False
    for i in range(3,int(n/2)+1, 2):   # 홀수만
        if prime[i] and prime[n - i]:
            print(f"{n} = {i} + {n - i}")
            flag = True
            break # 찾고나면 for문 종료
    if flag==False:
        print("Goldbach's conjecture is wrong.")
```
