## 소수 구하기

```python
# 에라토스테네스의 체 이용
import math

# input
m,n = map(int,input().split())

def is_prime(m,n):
    prime=[True for _ in range(n+1)]
    prime[1]=False
    for i in range(2,int(math.sqrt(n))+1):
        if prime[i]:
            for j in range(i*i,n+1,i): # i*i에서 시작 -> i보다 작은 수를 곱하는 연산은 이미 이전에 반복했으므로 중복 제거
                prime[j]=False

    for i in range(m,n+1):
        if prime[i]:
            print(i)

is_prime(m,n)
```
