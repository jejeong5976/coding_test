## 소수 찾기

```python
import math

# input
n = int(input())
data = map(int,input().split())

def is_prime(a):
    if a<2:
        return False
    for i in range(2,int(math.sqrt(a))+1):
        if a%i==0:
            return False
    return True

count=0
for a in data:
    if is_prime(a)==True:
        count+=1
print(count)
```

## 에라토스테네스의 체
- 하나씩 확인하는 방법은 비효율적 -> 시간 복잡도 O(n)
- 범위를 제곱근으로 줄여 확인하는 방식 -> 시간 복잡도는 O(sqrt(n))으로 줄어들지만 여전히 큰 수에 대해서는 비효율적일 수 있음
- 에라토스테네스의 체 -> 시간 복잡도 O(nlong(logn))
    - n까지의 모든 수를 나열해놓고 2,3,5,7... 이런 식으로 소수의 배수들을 체로 거르는 방식
    - 특정 숫자 범위 내의 모든 소수를 찾을 때 효율적 -> 소수 구하는 알고리즘에서 시간 초과가 나올 경우 참고
    1) 1부터 n을 원소로 갖는 배열 arr 선언
    2) 2부터 sqrt(n)까지 나누어 떨어지면 해당 배열의 원소를 삭제
    3) 이 때 나누어 떨어지는 수는 자기자신은 제외할 것

```python
import math

# 선언
MAX=100
prime=[True for i in range(MAX+1)]

# 체로 거르기
prime[1]=False # 1은 소수도, 합성수도 아니므로 false

for i in range(2,int(math.sqrt(MAX))+1):
    if prime[i]==True:
        j=2
        while i*j<=MAX:
            prime[i*j]=False #배수인 원소 false로 변환
            j+=1

for i in range(2,MAX+1):
    if prime[i]:
        print(i,end=' ')
```
