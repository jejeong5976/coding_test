## 팩토리얼 0의 개수

math.factorial 이용 시와 아래의 함수와 메모리와 시간은 동일.

```python
def factorial(n):
    if n<=1:
        return 1
    else:
        return n*factorial(n-1)

n=int(input())
str_answer=str(factorial(n))
count = 0
for s in str_answer[::-1]:
    if s=='0':
        count+=1
    else:
        break

print(count)
```
